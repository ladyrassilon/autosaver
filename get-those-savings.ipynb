```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import random
import os
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from tqdm import notebook
```

```python
browser = webdriver.Firefox()
```

```python
browser.get("https://www.instagram.com/")
```

## Manual Steps
- Click allow "Allow all cookies"
- Login to your instagram
- (you can save, or not, unfortunatly the browser isn't saving state at the moment)
- continue...

```python
username_files = [uf for uf in os.listdir("potential_savings/") if "usernames" in uf]
username_files
```

```python
lines = []
for uf in username_files:
    with open(f"potential_savings/{uf}", "r") as u_file:
        lines.extend(u_file.readlines())
```

```python
removed_endlines = [line.strip() for line in lines]
removed_endlines
```

```python
unique_users = list(set(removed_endlines))
unique_users
problems = []
```

```python
for unique_user in notebook.tqdm(unique_users):
    user_url = f"https://www.instagram.com/{unique_user}/"
    browser.get(user_url)
    time.sleep(2)
    problem_flag = False
    following = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Following')]") if el.is_displayed()]
    if following:
        print(f"removing old savings - {unique_user}")
        following[0].click()
        time.sleep(2)
        unfollow = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Unfollow')]") if el.is_displayed()]
        unfollow[0].click()
        time.sleep(2)
        try:
            dismiss = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Dismiss')]") if el.is_displayed()]
            dismiss[0].click()
        except IndexError:
            pass
    time.sleep(2)
    unblock = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Unblock')]") if el.is_displayed()]
    if not unblock:
        options_circles = [el for el in browser.find_elements(By.TAG_NAME, "circle") if el.is_displayed() and sum(el.size.values()) == 8]
        if options_circles:
            print(f"applying savings - {unique_user}")
            options_circle = options_circles[0]
            
            action = webdriver.common.action_chains.ActionChains(browser)
            action.move_to_element(options_circle)
            action.click()
            action.perform()
            time.sleep(2)
        
            block = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Block')]") if el.is_displayed()]
            block_button = block[0]
            block_button.click()
            time.sleep(2)
        
            block = [el for el in browser.find_elements(By.XPATH, "//button[contains(text(), 'Block')]") if el.is_displayed()]
            block_button = block[1]
            
            action = webdriver.common.action_chains.ActionChains(browser)
            action.move_to_element_with_offset(block_button, -10, -10)
            action.click()
            action.perform()
            time.sleep(2)
            try:
                dismiss = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Dismiss')]") if el.is_displayed()]
                dismiss[0].click()
            except IndexError:
                pass
        else:
            print(f"problem applying savings - {unique_user}")
            problems.append(unique_user)
            problem_flag = True
    else:
        print(f"savings already applied - {unique_user}")
    if following or not unblock or problem_flag:
        random_pause = random.randrange(30, 60)
        time.sleep(random_pause)
print("Script has finished running! Enjoy the savings")
if problems:
    print(f"The following {len(problems)} savings were not applied, re-run this cell to get them\n")
    print("\n - ".join(problems))
```
