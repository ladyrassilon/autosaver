```python
username_files = [uf for uf in os.listdir("potential_savings/") if "usernames" in uf]
username_files
```

```python
lines = []
for uf in username_files:
    with open(f"potential_savings/{uf}", "r") as u_file:
        lines.extend(u_file.readlines())
```

```python
removed_endlines = [line.strip() for line in lines]
removed_endlines
```

```python
unique_users = list(set(removed_endlines))
# unique_users.sort()
unique_users
# unique_users
```

```python
for unique_user in notebook.tqdm(unique_users):
    print(f"blocking - {unique_user}")
    user_url = f"https://www.instagram.com/{unique_user}/"
    browser.get(user_url)
    time.sleep(2)
    following = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Following')]") if el.is_displayed()]
    if following:
        following[0].click()
        time.sleep(2)
        unfollow = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Unfollow')]") if el.is_displayed()]
        unfollow[0].click()
        time.sleep(2)
        try:
            dismiss = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Dismiss')]") if el.is_displayed()]
            dismiss[0].click()
        except IndexError:
            pass
    time.sleep(2)
    unblock = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Unblock')]") if el.is_displayed()]
    if not unblock:
        options_circles = [el for el in browser.find_elements(By.TAG_NAME, "circle") if el.is_displayed() and sum(el.size.values()) == 8]
        options_circle = options_circles[0]
        
        action = webdriver.common.action_chains.ActionChains(browser)
        action.move_to_element(options_circle)
        action.click()
        action.perform()
        time.sleep(2)
    
        block = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Block')]") if el.is_displayed()]
        block_button = block[0]
        block_button.click()
        time.sleep(2)
    
        block = [el for el in browser.find_elements(By.XPATH, "//button[contains(text(), 'Block')]") if el.is_displayed()]
        block_button = block[1]
        
        action = webdriver.common.action_chains.ActionChains(browser)
        action.move_to_element_with_offset(block_button, -10, -10)
        action.click()
        action.perform()
        time.sleep(2)
        try:
            dismiss = [el for el in browser.find_elements(By.XPATH, "//*[contains(text(), 'Dismiss')]") if el.is_displayed()]
            dismiss[0].click()
        except IndexError:
            pass
    if following or not unblock:
        random_pause = random.randrange(30, 60)
        time.sleep(random_pause)
            # browser.find_elements(By.XPATH, "//*[contains(text(), 'Dismiss')]")[0].click()
```
