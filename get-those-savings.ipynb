```python
username_files = [uf for uf in os.listdir("potential_savings/") if "usernames" in uf]
username_files
```

```python
lines = []
for uf in username_files:
    with open(f"potential_savings/{uf}", "r") as u_file:
        lines.extend(u_file.readlines())
```
